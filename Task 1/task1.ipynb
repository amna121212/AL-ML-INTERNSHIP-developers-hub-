#  Exploring and Visualizing the Iris Dataset

```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Iris dataset
iris = sns.load_dataset('iris')
```

## Dataset Inspection

```python
# Print shape of dataset
print("Shape of Dataset:")
print(iris.shape)
```

```python
# Print column names
print("\nColumn Names:")
print(iris.columns)
```

```python
# Display first 5 rows
print("\nFirst 5 Rows:")
print(iris.head())
```

```python
# Dataset information
print("\nDataset Info:")
print(iris.info())
```

```python
# Summary statistics
print("\nSummary Statistics:")
print(iris.describe())
```

## Data Visualization

```python
# Set seaborn style
sns.set(style="whitegrid")
```

```python
# 1. Scatter Plot
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=iris,
    x='sepal_length',
    y='sepal_width',
    hue='species'
)
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.show()
```

```python
# 2. Histograms
iris.hist(figsize=(10,8))
plt.suptitle('Histograms of Iris Features')
plt.show()

# 3. Box Plots
plt.figure(figsize=(10,6))
sns.boxplot(data=iris)
plt.title('Box Plot of Iris Dataset Features')
plt.show()
```
